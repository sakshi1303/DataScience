```python
stock = pd.read_parquet('nse_data')
```

```python
stock.columns
```

```python
df_stock = stock[[' DATE1', 'SYMBOL', ' CLOSE_PRICE']]
df_stock.columns = ['Date', 'Symbol','Close Price']
df_stock.info()
```

```python
df_stock['Date'][0]
```

```python
df_stock['Date'] = df_stock['Date'].astype(str)
df_stock.info()
```

```python
df_stock['Date'] = df_stock['Date'].replace(' ', '', regex=True)
```

```python
df_stock['Date'][0]
```

```python
df_stock.info()
```

```python
df_stock.head()
```

```python
df_stock['Date'] = pd.to_datetime(df_stock['Date'],format='%d-%b-%Y')
```

```python
df_stock['year'] = df_stock.Date.dt.year
df_stock['year'] = pd.to_datetime(df_stock['year'],format='%Y')
```

```python
df_stock.info()
```

```python
df_stock.head()
```

```python
#df_stock.pivot(index='year', columns='Symbol', values='Close Price')

df_tmp = df_stock[['Date', 'Symbol', 'Close Price']]
#df1.set_index('year', inplace=True)
df_tmp.head()
```

```python
df_tmp.head()
```

```python
df_tmp.tail()
```

```python
df_tmp = df_tmp.sort_values(by='Date')
```

```python
df_tmp.tail()
```

```python
df_tmp.groupby('Symbol')['Close Price'].max()
```

```python
df_stocks = df_tmp.groupby(['Date', 'Symbol'])['Close Price'].max().unstack().fillna(method='ffill')
df_stocks = df_stocks.fillna(0)
```

```python
df_stocks
```

```python
top25_stocks = df_tmp.groupby(['Symbol'])[['Close Price']].max().sort_values(by='Close Price', ascending=False)[:25]
top25_stocks.index
```

```python
top25_stocks.head()
```

```python
df_top25_stocks = df_stocks[top25_stocks.index]
```

```python
df_top25_stocks
```
